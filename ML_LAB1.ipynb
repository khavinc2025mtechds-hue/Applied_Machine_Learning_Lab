```python
if 'price_range' in df.columns:
  for col in num_col:
    if col!='price_range':
      plt.figure(figsize=(8,6))
      sns.boxplot(x="price_range", y=col, data=df)
      plt.show()
      plt.figure(figsize=(8,6))
      sns.violinplot(x="price_range", y=col, data=df)
      plt.show()
```
